```python
# Cargar el JSON CatBoost
with open('results_catboost_2.json', 'r') as f:
    data_CB = json.load(f)

# Extraer métricas de entrenamiento CatBoost
train_MSE_CB= data_CB['mse_train']
train_RMSE_CB= data_CB['rmse_train']
train_R2_CB= data_CB['r2_train']
train_MAE_CB= data_CB['mae_train']
# Extraer métricas de test CatBoost
test_MSE_CB= data_CB['mse_test']
test_RMSE_CB= data_CB['rmse_test']
test_R2_CB= data_CB['r2_test']
test_MAE_CB= data_CB['mae_test']


# Extraer hiperparámetros CatBoost
best_params_CB = data_CB["best_params"]
best_params_CB_bootstrap_type = best_params_CB['bootstrap_type']
best_params_CB_iterations = best_params_CB['iterations']
best_params_CB_depth = best_params_CB['depth']
best_params_CB_learning_rate= best_params_CB['learning_rate']
best_params_CB_l2_leaf_reg = best_params_CB['l2_leaf_reg']
best_params_CB_random_strength = best_params_CB['random_strength']
best_params_CB_border_count = best_params_CB['border_count']
best_params_CB_min_data_in_leaf = best_params_CB['min_data_in_leaf']
best_params_CB_loss_function = best_params_CB['loss_function']
best_params_CB_od_type = best_params_CB['od_type']
best_params_CB_od_wait = best_params_CB['od_wait']
best_params_CB_subsample = best_params_CB['subsample']
```

```python
# Abrimos experimento en MLFlow
mlflow.set_experiment('Análisis de estrés')
# Iniciamos run modelo LSTM
with mlflow.start_run(run_name='Modelo_LSTM'):
    # Guardamos las métricas de entrenamiento de LSTM
    mlflow.log_metric("train_MSE", train_MSE_LSTM)
    mlflow.log_metric("train_RMSE", train_RMSE_LSTM)
    mlflow.log_metric("train_R2", train_R2_LSTM)
    mlflow.log_metric("train_MAE", train_MAE_LSTM)
    mlflow.log_metric("train_Loss", train_Loss_LSTM)
    mlflow.log_metric("val_MSE", val_MSE_LSTM)
    mlflow.log_metric("val_RMSE", val_RMSE_LSTM)
    mlflow.log_metric("val_R2", val_R2_LSTM)
    mlflow.log_metric("val_MAE", val_MAE_LSTM)
    mlflow.log_metric("val_Loss", val_Loss_LSTM)
    mlflow.log_metric("test_MSE", test_MSE_LSTM)
    mlflow.log_metric("test_RMSE", test_RMSE_LSTM)
    mlflow.log_metric("test_R2", test_R2_LSTM)
    mlflow.log_metric("test_MAE", test_MAE_LSTM)
    mlflow.log_metric("test_Loss", test_Loss_LSTM)
    # Guardamos los mejores parámetros
    mlflow.log_param("hidden_dim", best_params_hidden_dim_LSTM)
    mlflow.log_param("num_layers", best_params_num_layers_LSTM)
    mlflow.log_param("dropout", best_params_dropout_LSTM)
    mlflow.log_param("optimizer", best_params_optimezer_LSTM)
    mlflow.log_param("lr", best_params_learning_rate_LSTM)
    mlflow.log_param("criterion", best_params_criterion_LSTM)
    # Guardamos el modelo
    mlflow.log_artifact("metricas_y_best_params_LSTM.json")
    # Guardamos el .txt
    mlflow.log_artifact("LSTM_X_train_columns.txt")
    
# Iniciamos run modelo Random Forest
with mlflow.start_run(run_name='Modelo_Random_Forest'):
    # Guardamos las métricas
    mlflow.log_metric("train_accuracy", train_accuracy_RF)
    mlflow.log_metric("train_f1_score", train_f1_score_RF)
    mlflow.log_metric("train_precision", train_precision_RF)
    mlflow.log_metric("train_recall", train_recall_RF)

    mlflow.log_metric("test_accuracy", test_accuracy_RF)
    mlflow.log_metric("test_f1_score", test_f1_score_RF)
    mlflow.log_metric("test_precision", test_precision_RF)
    mlflow.log_metric("test_recall", test_recall_RF)

    # Guardamos los mejores parámetros
    mlflow.log_param("n_estimators", best_params_RF_n_estimators)
    mlflow.log_param("criterion", best_params_RF_criterion)
    mlflow.log_param("max_depth", best_params_RF_max_depth)
    mlflow.log_param("min_samples_split", best_params_RF_min_samples_split)
    mlflow.log_param("min_samples_leaf", best_params_RF_min_samples_leaf)
    mlflow.log_param("min_weight_fraction_leaf", best_params_RF_min_weight_fraction_leaf)
    mlflow.log_param("max_features", best_params_RF_max_features)
    mlflow.log_param("max_leaf_nodes", best_params_RF_max_leaf_nodes)
    mlflow.log_param("min_impurity_decrease", best_params_RF_min_impurity_decrease)
    mlflow.log_param("class_weight", best_params_RF_class_weight)
    mlflow.log_param("ccp_alpha", best_params_RF_ccp_alpha)
    mlflow.log_param("max_samples", best_params_RF_max_samples)
    mlflow.log_param("oob_score", best_params_RF_oob_score)
    # Guardamos el modelo
    mlflow.log_artifact("metricas_y_best_params_Random_Forest.json")
    # Guardamos el .txt
    mlflow.log_artifact("columnas_X_train_Random_Forest.txt")

# Iniciamos run modelo XGBoost_1
with mlflow.start_run(run_name='Modelo_XGBoost_2'):
    # Guardamos las métricas
    mlflow.log_metric("train_MSE", train_MSE_XGB)
    mlflow.log_metric("train_RMSE", train_RMSE_XGB)
    mlflow.log_metric("train_R2", train_R2_XGB)
    mlflow.log_metric("train_MAE", train_MAE_XGB)
    mlflow.log_metric("test_MSE", test_MSE_XGB)
    mlflow.log_metric("test_RMSE", test_RMSE_XGB)
    mlflow.log_metric("test_R2", test_R2_XGB)
    mlflow.log_metric("test_MAE", test_MAE_XGB)
   
    # Guardamos los mejores parámetros
    mlflow.log_param("n_estimators", best_params_XGBoost_1_n_estimators)
    mlflow.log_param("max_depth", best_params_XGBoost_1_max_depth)
    mlflow.log_param("learning_rate", best_params_XGBoost_1_learning_rate)
    mlflow.log_param("subsample", best_params_XGBoost_1_subsample)
    mlflow.log_param("colsample_bytree", best_params_XGBoost_1_colsample_bytree)
    mlflow.log_param("reg_lambda", best_params_XGBoost_1_reg_lambda)
    mlflow.log_param("reg_alpha", best_params_XGBoost_1_reg_alpha)
    mlflow.log_param("min_child_weight", best_params_XGBoost_1_min_child_weight)
    mlflow.log_param("gamma", best_params_XGBoost_1_gamma)
    # Guardamos el modelo
    mlflow.log_artifact("results_xgb_stress_2.json")
    # Guardamos el .txt
    mlflow.log_artifact("XGboost_columns_x_train.txt")

# Iniciamos run Decision Tree
with mlflow.start_run(run_name='Modelo_Decision_Tree'):
    # Guardamos las métricas
    mlflow.log_metric("train_accuracy", train_accuracy_DT)
    mlflow.log_metric("train_f1_score", train_f1_score_DT)
    mlflow.log_metric("train_precision", train_precision_DT)
    mlflow.log_metric("train_recall", train_recall_DT)

    mlflow.log_metric("test_accuracy", test_accuracy_DT)
    mlflow.log_metric("test_f1_score", test_f1_score_DT)
    mlflow.log_metric("test_precision", test_precision_DT)
    mlflow.log_metric("test_recall", test_recall_DT)


    # Guardamos los mejores parámetros
    mlflow.log_param("max_depth", best_params_DT_max_depth)
    mlflow.log_param("min_samples_leaf", best_params_DT_min_samples_leaf)
    mlflow.log_param("max_leaf_nodes", best_params_DT_max_leaf_nodes)
    mlflow.log_param("min_impurity_decrease", best_params_DT_min_impurity_decrease)
    
    # Guardamos el modelo
    mlflow.log_artifact("metricas_y_best_params_DecisionTree.json")
    # Guardamos el .txt
    mlflow.log_artifact("columnas_X_train_DecisionTree.txt")
    
# Iniciamos run Regresiom Logística
with mlflow.start_run(run_name='Modelo_Regresión_Logística'):
    # Guardamos las métricas
    mlflow.log_metric("train_accuracy", train_accuracy_RL)
    mlflow.log_metric("train_f1_score", train_f1_score_RL)
    mlflow.log_metric("train_precision", train_precision_RL)
    mlflow.log_metric("train_recall", train_recall_RL)

    mlflow.log_metric("test_accuracy", test_accuracy_RL)
    mlflow.log_metric("test_f1_score", test_f1_score_RL)
    mlflow.log_metric("test_precision", test_precision_RL)
    mlflow.log_metric("test_recall", test_recall_RL)


    # Guardamos los mejores parámetros
    mlflow.log_param("penalty", best_params_RL_penalty)
    mlflow.log_param("solver", best_params_RL_solver)
    mlflow.log_param("max_iter", best_params_RL_max_iter)
    
    
    # Guardamos el modelo
    mlflow.log_artifact("metricas_y_best_params_RegresiónLogística.json")
    # Guardamos el .txt
    mlflow.log_artifact("columnas_X_train_RegresiónLogística.txt")

# Iniciamos run CatBoost
with mlflow.start_run(run_name='Modelo_CatBoost'):
    # Guardamos las métricas
    mlflow.log_metric("train_MSE", train_MSE_CB)
    mlflow.log_metric("train_RMSE", train_RMSE_CB)
    mlflow.log_metric("train_R2", train_R2_CB)
    mlflow.log_metric("train_MAE", train_MAE_CB)
    mlflow.log_metric("test_MSE", test_MSE_CB)
    mlflow.log_metric("test_RMSE", test_RMSE_CB)
    mlflow.log_metric("test_R2", test_R2_CB)
    mlflow.log_metric("test_MAE", test_MAE_CB)

    # Guardamos los mejores parámetros
    mlflow.log_param("bootstrap_type", best_params_CB_bootstrap_type )
    mlflow.log_param("iterations", best_params_CB_iterations)
    mlflow.log_param("depth", best_params_CB_depth)
    mlflow.log_param("learning_rate", best_params_CB_learning_rate)
    mlflow.log_param("l2_leaf_reg", best_params_CB_l2_leaf_reg)
    mlflow.log_param("random_strength", best_params_CB_random_strength)
    mlflow.log_param("border_count", best_params_CB_border_count)
    mlflow.log_param("min_data_in_leaf", best_params_CB_min_data_in_leaf)
    mlflow.log_param("loss_function", best_params_CB_loss_function)
    mlflow.log_param("od_type", best_params_CB_od_type)
    mlflow.log_param("od_wait", best_params_CB_od_wait)
    mlflow.log_param("subsample", best_params_CB_subsample)

    # Guardamos el modelo
    mlflow.log_artifact("results_catboost_2.json")
    

```
