```python
import pandas as pd
import numpy as np
```

```python
def read_data(filename):
    return pd.read_csv(filename, on_bad_lines='skip')
```

```python
data_1 = read_data("../Dataset/C - Body posture features 1.csv")
data_4 = read_data("../Dataset/C - Body posture features 4.csv")

data_1 = data_1.drop_duplicates()
data_4 = data_4.drop_duplicates()

del data_4['Unnamed: 2'] 

cols_1, cols_4 = list(data_1), list(data_4)

```

```python
len(cols_1), len(cols_4)
```

```python
data_1 = data_1.sort_values('timestamp')
data_4 = data_4.sort_values('timestamp')
```

```python
#data_1.to_csv('../Dataset/generated/C_1.csv', index=False) 
#data_4.to_csv('../Dataset/generated/C_4.csv', index=False) 
```

```python
data_1.shape, data_4.shape
```

```python
import matplotlib.pyplot as plt

N = 3000

for feature in list(data_1)[3:30]:
    x = data_1[feature][:N].values.tolist()
    y = data_4[feature][:N].values.tolist()
    x = [float(i) for i in x]

    fig = plt.figure()

    plt.xlim(min(x), max(x))
    plt.ylim(min(y), max(y))
    plt.scatter(x, y, s=25, alpha=0.5)
    fig.text(.5, .0001, feature, ha='center')
    plt.show()
```
