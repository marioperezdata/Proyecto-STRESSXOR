# 0. Descripción de los subsets del grupo C

Este subset decribe la posición corporal de las personas involucradas en el experimento.

+ C - Body posture features (Kinect - final (annotated and selected))-sheet_1 y C - Body posture features (Kinect - final (annotated and selected))-sheet_4 :
En este archivo se recopilan las features que se han generado a través de del reconocimiento de las posturas corporales de los sujetos en el experimento. Se usa el sensor Kinect 3D, extrayendo las features con Kinect SDK. Hay varias columnas que se detallan a continuación:

  *Metadatos generales:*
  - **PP**: Identificador del participante.
  - **C**: Condición experimental en la que se realizó la medición.
  - **timestamp**: Marca de tiempo indicando el minuto exacto de la medición.

  *Distancia:*
  - **avgDepth(avg y stdv)**: Profundidad promedio del participante en la escena, medida por el sensor.

  *Ángulos articulares (Joint Angles):*
  - **leftShoulderAngle(avg y stdv)**: Ángulo promedio del hombro izquierdo.
  - **rightShoulderAngle(avg y stdv)**: Ángulo promedio del hombro derecho.
  - **leanAngle(avg y stdv)**: Ángulo de inclinación promedio del torso.
  - **HipCenter_Spine-Spine_ShoulderCenter(avg y stdv)**: Ángulo entre la cadera y la parte superior de la columna.
  - **Spine_ShoulderCenter-ShoulderCenter_Head(avg y stdv)**: Ángulo entre la parte superior de la columna y la cabeza.
  - **Spine_ShoulderCenter-ShoulderCenter_ShoulderLeft(avg y stdv)**: Ángulo entre la parte superior de la columna y el hombro izquierdo.
  - **Spine_ShoulderCenter-ShoulderCenter_ShoulderRight(avg y stdv)**: Ángulo entre la parte superior de la columna y el hombro derecho.
  - **ShoulderCenter_ShoulderLeft-ShoulderLeft_ElbowLeft(avg y stdv)**: Ángulo entre el hombro izquierdo y el codo izquierdo.
  - **ShoulderLeft_ElbowLeft-ElbowLeft_WristLeft(avg y stdv)**: Ángulo entre el codo izquierdo y la muñeca izquierda.
  - **ShoulderCenter_ShoulderRight-ShoulderRight_ElbowRight(avg y stdv)**: Ángulo entre el hombro derecho y el codo derecho.
  - **ElbowLeft_WristLeft-WristLeft_HandLeft(avg y stdv)**: Ángulo entre el codo izquierdo, la muñeca izquierda y la mano izquierda.
  - **ShoulderCenter_ShoulderRight-ShoulderRight_ElbowRight(avg y stdv)**: Ángulo entre el centro de los hombros, el hombro derecho y el codo derecho.
  - **ShoulderRight_ElbowRight-ElbowRight_WristRight(avg y stdv)**: Ángulo entre el hombro derecho, el codo derecho y la muñeca derecha.
  - **ElbowRight_WristRight-WristRight_HandRight(avg y stdv)**: Ángulo entre el codo derecho, la muñeca derecha y la mano derecha.

  *Orientaciones óseas (Bone Orientations) en tres planos (ZX, XY, YZ):*
  - **HipCenter_Spine-PlaneZXAxisX(avg y stdv)**: Orientación del hueso entre la cadera y la columna en el plano ZX.
  - **HipCenter_Spine-PlaneXYAxisY(avg y stdv)**: Orientación del hueso entre la cadera y la columna en el plano XY.
  - **HipCenter_Spine-PlaneYZAxisZ(avg y stdv)**: Orientación del hueso entre la cadera y la columna en el plano YZ.
  - **Spine_ShoulderCenter-PlaneZXAxisX(avg y stdv)**: Orientación del hueso entre la parte superior de la columna y el centro de los hombros en el plano ZX.
  - **Spine_ShoulderCenter-PlaneXYAxisY(avg y stdv)**: Orientación del hueso entre la parte superior de la columna y el centro de los hombros en el plano XY.
  - **Spine_ShoulderCenter-PlaneYZAxisZ(avg y stdv)**: Orientación del hueso entre la parte superior de la columna y el centro de los hombros en el plano YZ.
  - **ShoulderCenter_Head-PlaneZXAxisX(avg y stdv)**: Orientación del hueso entre el centro de los hombros y la cabeza en el plano ZX.
  - **ShoulderCenter_Head-PlaneXYAxisY(avg y stdv)**: Orientación del hueso entre el centro de los hombros y la cabeza en el plano XY.
  - **ShoulderCenter_Head-PlaneYZAxisZ(avg y stdv)**: Orientación del hueso entre el centro de los hombros y la cabeza en el plano YZ.
  - **ShoulderCenter_ShoulderLeft-PlaneZXAxisX(avg y stdv)**: Orientación del hueso entre el centro de los hombros y el hombro izquierdo en el plano ZX.
  - **ShoulderCenter_ShoulderLeft-PlaneXYAxisY(avg y stdv)**: Orientación del hueso entre el centro de los hombros y el hombro izquierdo en el plano XY.

  - **ShoulderCenter_ShoulderRight-PlaneXYAxisY(avg y stdv)**: Orientación del hueso entre el centro de los hombros y el hombro derecho en el plano XY.

  - **ShoulderLeft_ElbowLeft-PlaneZXAxisX(avg y stdv)**: Orientación del hueso entre el hombro izquierdo y el codo izquierdo en el plano ZX.
  - **ShoulderLeft_ElbowLeft-PlaneXYAxisY(avg y stdv)**: Orientación del hueso entre el hombro izquierdo y el codo izquierdo en el plano XY.
